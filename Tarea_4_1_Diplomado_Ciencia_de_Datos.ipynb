```python
import pandas as pd

class IrisDatasetIterator:
    def __init__(self, filename):
        self.dataframe = pd.read_csv(filename)
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index < len(self.dataframe):
            row1 = self.dataframe.iloc[self.index]
            row2 = self.dataframe.iloc[self.index + 1]
            self.index += 2
            return row1['sepal.length'], row1['sepal.width'], row2['sepal.length'], row2['sepal.width']
        else:
            raise StopIteration("error")

dataset_iterator = IrisDatasetIterator("iris.csv")
```

```python
#PRUEBA DE LA CLASE IrisDatasetIterator
dataset_iterator = IrisDatasetIterator("iris.csv")

for _ in range(round(len(dataset_iterator.dataframe)/2) + 1):
    try:
      a, b, c, d = next(dataset_iterator)
      print(f"Sepal Length: {a}, Sepal Width: {b} \n Sepal Length: {c}, Sepal Width: {d}")
    except StopIteration as e:
      print(e)
```
