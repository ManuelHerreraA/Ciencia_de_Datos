```python
import pandas as pd
import matplotlib.pyplot as plt

class DescriptiveStatistics:
    def __init__(self, filename):
        self.dataframe = pd.read_csv(filename)

    def calculate_basic_stats(self, variable_name):
        if variable_name in self.dataframe.columns:
            variable = self.dataframe[variable_name]
            mean = variable.mean()
            median = variable.median()
            std_dev = variable.std()
            q25 = variable.quantile(0.25)
            q50 = variable.quantile(0.50)
            q75 = variable.quantile(0.75)
            return {
                'Mean': mean,
                'Median': median,
                'Standard Deviation': std_dev,
                'Q25': q25,
                'Q50': q50,
                'Q75': q75
            }
        else:
            return None

    def available_variables(self):
        return self.dataframe.columns.tolist()

    def plot_histogram(self, variable_name):
        if variable_name in self.dataframe.columns:
            variable = self.dataframe[variable_name]
            plt.hist(variable, bins=20, edgecolor='black')
            plt.title(f'Histogram of {variable_name}')
            plt.xlabel(variable_name)
            plt.ylabel('Frequency')
            plt.show()
```

```python
# PRUEBA DE LA CLASE DescriptiveStatistics

stats = DescriptiveStatistics("iris.csv")

variable_names = stats.available_variables()
print("Available variables:", variable_names)

for i in variable_names[:-1]:
  variable_stats = stats.calculate_basic_stats(i)
  if variable_stats:
      print(f"Basic statistics for {i}:")
      for stat, value in variable_stats.items():
          print(f"{stat}: {value}")
      stats.plot_histogram(i)
  else:
      print(f"Variable '{i}' not found in the dataset.")
```
